### Banco de dados db_imoveis_sp

```python
#mudando para o banco de dados db_imoveis_sp
db = client["db_imoveis_sp"]
```

```python
#selecionando a coleção imoveis_sp
colecao = db["imoveis_sp"]
```

```python
#verificando os documentos presentes na coleção
show_collection(colecao)
```

```python
#filtrando todos os documentos com bairro "paraiso"
resultado = colecao.find({"bairro": "paraiso"})
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com bairro "vila-mariana"
resultado = colecao.find({"bairro": "vila-mariana"})
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com área maior que 60, ordenados pela área em ordem crescente, limitados aos 15 primeiros documentos
resultado = colecao.find({"areaM2": {"$gt": 60}}).sort("areaM2",1).limit(15)
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com área igual a 60, ordenados pela área em ordem crescente, limitados aos 15 primeiros documentos
resultado = colecao.find({"areaM2":  60}).sort("areaM2",1)
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com área maior ou igual a 60, ordenados pela área em ordem crescente
resultado = colecao.find({"areaM2": {"$gte": 60}}).sort("areaM2",1)
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com área igual a 40 ou 60, ordenados pela área em ordem crescente
resultado = colecao.find({"areaM2": {"$in": [40, 60]}}).sort("areaM2",1)
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com área menor que 60, ordenados pela área em ordem decrescente
resultado = colecao.find({"areaM2": {"$lt": 60}}).limit(15).sort("areaM2",-1)
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com área igual ou menor a 60, ordenados pela área em ordem crescente limitados a 20 documentos
resultado = colecao.find({"areaM2": {"$lte": 60}}).sort("areaM2",-1).limit(20)
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com área diferente de 60, ordenados pela área em ordem crescente e limitados a 15 documentos
resultado = colecao.find({"areaM2": {"$ne": 60}}).sort("areaM2",1).limit(15)
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com área diferente de 20, 60 e 80, ordenados pela área em ordem crescente, limitados a 25 documentos
resultado = colecao.find({"areaM2": {"$in": [20, 60, 80]}}).sort("areaM2",1).limit(25)
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com $or ordenados em ordem decrescente
resultado = colecao.find({"$or": [{"vagas": 2}, {"suites": 2}]}, {"_id": 0}).sort("vagas",-1).limit(10)
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com $and ordenados em ordem decrescente
resultado = colecao.find({"$and": [{"vagas": 2}, {"suites": 2}]}, {"_id": 0}).sort("vagas",-1).limit(10)
for documento in resultado:
  print(documento)
```
